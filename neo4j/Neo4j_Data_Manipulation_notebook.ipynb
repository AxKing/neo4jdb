```python
def random_date():
    rand_year = np.random.randint(0,24)
    rand_year
    rand_month = np.random.randint(1,13)
    rand_month
    rand_day = np.random.randint(1,32)
    rand_day

    big_months = [1, 3, 5, 7, 8, 10, 12]
    feb = 2

    if rand_month == 2 and rand_day > 28:
        rand_day = 28
    if rand_month not in big_months and rand_day == 31:
        rand_day = 30
    if rand_year < 10:
        date = '200'+str(rand_year)+'-'+str(rand_month)+'-'+str(rand_day)
    else: 
        date = '20'+str(rand_year)+'-'+str(rand_month)+'-'+str(rand_day)
    return date
```

```python
date_col = []
for i in range(len(df)):
    date_col.append(random_date())
print(date_col)
```

```python
df['document_date'] = date_col
```
